```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nclass = 500 # Number of instances in each class
```

```python
X1 = np.random.randn(Nclass,2) + np.array([0,-2]) # mean = 0  ; variance = 2
X2 = np.random.randn(Nclass,2) + np.array([2,2])
X3 = np.random.randn(Nclass,2) + np.array([-2,2])
X = np.concatenate((X1,X2,X3),axis=0)  # Input
Y = np.array([0]*Nclass + [1]*Nclass + [2]*Nclass) #1500 values (Targets)
```

```python
X.shape # N x D
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=100,alpha=0.5)
plt.show()

```

```python
D = 2 #Inputs
M = 3 #Hidden nodes
K = 3 #Outputs (3 classes)
```

```python
W1 = np.random.randn(D,M) 
b1 = np.random.randn(M)
W2 = np.random.randn(M,K)
b2 = np.random.randn(K)
```

```python
def forward(X,W1,b1,W2,b2):
    Z = 1/(1+np.exp(-X.dot(W1)-b1))
    A = Z.dot(W2)+b2
    expa = np.exp(A)
    Y = expa/expa.sum(axis=1,keepdims=True) #See softmax.ipynb
    return Y
```

```python
def classification(Y,P):
    n_correct =0
    n_total =0
    for i in range(len(Y)):
        n_total+=1
        if(Y[i]==P[i]):
            n_correct+=1
    return float(n_correct)/n_total
```

```python
P_Y_given_X = forward(X,W1,b1,W2,b2)
```

```python
# np.argmax() will find highest value (highest probability) in a row and obtain the position of it in the row.
P = np.argmax(P_Y_given_X,axis=1) 
```

```python
assert(len(P)==len(Y)) #No output means true
```

```python
print("Classification accuracy : "+str(classification(Y,P)))
```
