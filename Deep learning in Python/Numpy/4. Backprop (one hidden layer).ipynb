```python
#Copied from forward prop

import numpy as np
import matplotlib.pyplot as plt

def forward(X,W1,b1,W2,b2):
    Z = 1/(1+np.exp(-X.dot(W1)-b1))
    A = Z.dot(W2)+b2
    expa = np.exp(A)
    Y = expa/expa.sum(axis=1,keepdims=True) #See softmax.ipynb
    return Y,Z #### Minor Change

def classification(Y,P):
    n_correct =0
    n_total =0
    for i in range(len(Y)):
        n_total+=1
        if(Y[i]==P[i]):
            n_correct+=1
    return float(n_correct)/n_total

Nclass = 500
D = 2 #Inputs
M = 3 #Hidden nodes
K = 3 #Outputs (3 classes)
X1 = np.random.randn(Nclass,2) + np.array([0,-2]) # mean = 0  ; variance = 2
X2 = np.random.randn(Nclass,2) + np.array([2,2])
X3 = np.random.randn(Nclass,2) + np.array([-2,2])
X = np.concatenate((X1,X2,X3),axis=0)  # Input
Y = np.array([0]*Nclass + [1]*Nclass + [2]*Nclass) #1500 values (Targets)

W1 = np.random.randn(D,M) 
b1 = np.random.randn(M)
W2 = np.random.randn(M,K)
b2 = np.random.randn(K)
```

```python
N = len(Y)
N
```

```python
T = np.zeros((N,K)) # Target values
```

```python
for i in range(N): # One hot encoding
    T[i,Y[i]]=1
```

```python
learning_rate = 10e-7
costs = []
```

```python
def cost(T,Y):
    tot = T * np.log(Y) # J
    return tot.sum()
    
```

```python
def derivative_w2(Z,T,Y): # T is target and Y is output obtained
    der = Z.T.dot(T-Y)
    return der
```

```python
def derivative_b2(T,Y):
    der = (T-Y)
    return der.sum(axis=0) #Always check dimensions of output to see if you need to sum over
```

```python
def derivative_w1(X,Z,T,Y,W2):
    dz = (T-Y).dot(W2.T)*Z*(1-Z)
    return (X.T).dot(dz)
```

```python
def derivative_b1(T,Y,W2,Z):
    dz = (T-Y).dot(W2.T)*Z*(1-Z)
    return dz.sum(axis=0) 
```

```python
for epoch in range(100000):
    output,hidden = forward(X,W1,b1,W2,b2)
    if(epoch%100==0):
        c = cost(T,output)
        P = np.argmax(output,axis=1)
        r = classification(Y,P)
        print("epoch : "+str(epoch)+"; Cost : "+str(c)+"; Accuracy : "+str(r))
        costs.append(c)
    #Gradient Ascent....(Maximising J [Same as Minimising L])
    W2 = W2 + learning_rate*derivative_w2(hidden,T,output)
    b2 = b2 + learning_rate*derivative_b2(T,output)
    W1 = W1 + learning_rate*derivative_w1(X,hidden,T,output,W2)
    b1 = b1 + learning_rate*derivative_b1(T,output,W2,hidden)
plt.plot(costs)
plt.show()
```

```python
#Binary

import numpy as np
import matplotlib.pyplot as plt

def forward(X,W1,b1,W2,b2):
    Z = 1/(1+np.exp(-X.dot(W1)-b1))
    A = Z.dot(W2)+b2
    expa = np.exp(A)
    Y = expa/expa.sum(axis=1,keepdims=True) #See softmax.ipynb
    return Y,Z #### Minor Change

def classification(Y,P):
    n_correct =0
    n_total =0
    for i in range(len(Y)):
        n_total+=1
        if(Y[i]==P[i]):
            n_correct+=1
    return float(n_correct)/n_total

Nclass = 500
D = 2 #Inputs
M = 3 #Hidden nodes
K = 1 #Outputs (2 classes)
X1 = np.random.randn(Nclass,2) + np.array([0,-2]) # mean = 0  ; variance = 2
X2 = np.random.randn(Nclass,2) + np.array([2,2])
X = np.concatenate((X1,X2),axis=0)  # Input
Y = np.array([0]*Nclass + [1]*Nclass) #1500 values (Targets)

W1 = np.random.randn(D,M) 
b1 = np.random.randn(M)
W2 = np.random.randn(M,K)
b2 = np.random.randn(K)

N = len(Y)
T = np.zeros((N,K+1))

for i in range(N): # One hot encoding
    T[i,Y[i]]=1
    
learning_rate = 10e-7
costs = []


for epoch in range(100000):
    output,hidden = forward(X,W1,b1,W2,b2)
    if(epoch%100==0):
        c = cost(T,output)
        P = np.argmax(output,axis=1)
        r = classification(Y,P)
        print("epoch : "+str(epoch)+"; Cost : "+str(c)+"; Accuracy : "+str(r))
        costs.append(c)
    #Gradient Ascent....(Maximising J [Same as Minimising L])
    W2 = W2 + learning_rate*derivative_w2(hidden,T,output)
    b2 = b2 + learning_rate*derivative_b2(T,output)
    W1 = W1 + learning_rate*derivative_w1(X,hidden,T,output,W2)
    b1 = b1 + learning_rate*derivative_b1(T,output,W2,hidden)
plt.plot(costs)
plt.show()
```
