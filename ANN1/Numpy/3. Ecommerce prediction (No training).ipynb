```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('ecommerce_data.csv')
```

```python
data
```

```python
data.iloc[:,1] = (data.iloc[:,1]-data.iloc[:,1].mean()) /data.iloc[:,1].std() #Normalise is_mobile and n_products_viewed
```

```python
data.iloc[:,2] = (data.iloc[:,2]-data.iloc[:,2].mean()) /data.iloc[:,2].std()
```

```python
one_hot = pd.get_dummies(data['time_of_day'])
data.drop('time_of_day',axis = 1,inplace=True)
data = data.join(one_hot)
```

```python
Y = data['user_action']
X = data.drop('user_action',axis=1)
Y = np.array(Y)
X = np.array(X)
```

```python
# randomly initialize weights
M = 5
D = X.shape[1]
K = len(set(Y))
W1 = np.random.randn(D, M)
b1 = np.zeros(M)
W2 = np.random.randn(M, K)
b2 = np.zeros(K)
```

```python
# make predictions
def softmax(a):
    expA = np.exp(a)
    return expA / expA.sum(axis=1, keepdims=True)

def forward(X, W1, b1, W2, b2):
    Z = np.tanh(X.dot(W1) + b1)
    return softmax(Z.dot(W2) + b2)

# calculate the accuracy
def classification_rate(Y, P):
    return np.mean(Y == P)
```

```python
P_Y_given_X = forward(X, W1, b1, W2, b2)
print("P_Y_given_X.shape:", P_Y_given_X.shape)
predictions = np.argmax(P_Y_given_X, axis=1)
print("Score:", classification_rate(Y, predictions))
```
